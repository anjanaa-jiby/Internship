```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/final_dataset.csv')

# Display the first 5 rows
df.head()
```

```python
# Check the column names and data types
print(df.dtypes)
```

```python
# Create a comparison table
comparison = pd.DataFrame({
    'Original Text': df['text'].head(),
    'Lowercase Text': df['text'].str.lower().head()
})

comparison
```

```python
# Tokenization without NLTK and without loops

df['tokens'] = df['text'].astype(str).str.split()

# Display first 5 samples
df[['text', 'tokens']].head()
```

```python
print(df.columns)
```

```python
import nltk
from nltk.corpus import stopwords

# Download stopwords (run once)
nltk.download('stopwords')

# Load English stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords from the tokenized text
df['tokens_no_stopwords'] = df['tokens'].apply(
    lambda words: [word for word in words if word.lower() not in stop_words]
)

# Display before and after results for first 5 samples
df[['tokens', 'tokens_no_stopwords']].head()
```

```python
import nltk
from nltk.stem import PorterStemmer

# Create Porter Stemmer object
ps = PorterStemmer()

# Apply stemming
df['stemmed_tokens'] = df['tokens_no_stopwords'].apply(
    lambda words: [ps.stem(word) for word in words]
)

# Display stemmed output for first 10 samples
df[['tokens_no_stopwords', 'stemmed_tokens']].head(10)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download WordNet resources (run once)
nltk.download('wordnet')
nltk.download('omw-1.4')

# Create lemmatizer object
lemmatizer = WordNetLemmatizer()

# Apply lemmatization
df['lemmatized_tokens'] = df['tokens_no_stopwords'].apply(
    lambda words: [lemmatizer.lemmatize(word) for word in words]
)

# Compare stemming and lemmatization results
df[['tokens_no_stopwords', 'stemmed_tokens', 'lemmatized_tokens']].head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Convert token lists back to sentences
corpus = df['lemmatized_tokens'].apply(lambda x: ' '.join(x))

# Create CountVectorizer object
vectorizer = CountVectorizer()

# Fit and transform the text data
X = vectorizer.fit_transform(corpus)

# Display Vocabulary
print("Vocabulary:")
print(vectorizer.vocabulary_)

# Display Feature Names
print("\nFeature Names:")
print(vectorizer.get_feature_names_out())

# Display BoW Matrix
print("\nBag of Words Matrix:")
print(X.toarray())

# Display Shape
print("\nShape of Matrix:")
print(X.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert lemmatized tokens back to sentences
corpus = df['lemmatized_tokens'].apply(lambda x: ' '.join(x))

# Create TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Display Vocabulary
print("Vocabulary:")
print(tfidf_vectorizer.vocabulary_)

# Display TF-IDF Matrix
print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())

# Display Shape
print("\nShape of TF-IDF Matrix:")
print(tfidf_matrix.shape)
```
