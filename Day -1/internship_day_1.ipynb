```python
cf['Age_mean_imputed']=cf['Age'].fillna(cf['Age'].mean())
cf['Age_median_imputed']=cf['Age'].fillna(cf['Age'].median())
cf['Cabin_mode_imputed']=cf['Cabin'].fillna(cf['Cabin'].mode()[0])
```

```python
cf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.boxplot(data=cf,x='Pclass',y='Age',palette='pastel')

```

```python
plt.title('Boxplot of Age by Passenger Class',fontsize=16)
plt.xlabel('Passenger Class',fontsize=12)
plt.ylabel('Age',fontsize=12)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.boxplot(data=cf,x='Pclass',y='Survived',palette='pastel')

```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.boxplot(data=cf,x='Embarked',y='Fare',palette='pastel')
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
encode_data=cf.copy()
label_encode=LabelEncoder()
encode_data['Sex_Label']=label_encode.fit_transform(encode_data['Sex'])


```

```python
one_hot_encoded = pd.get_dummies(encode_data['Embarked'], prefix='Embarked')
encode_data = pd.concat([encode_data, one_hot_encoded], axis=1)

# Since 'Pclass' is ordinal, we can create a mapped column for demonstration.
pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third'}
# 3. Ordinal Encoding For p class
encode_data['Pclass_Category'] = encode_data['Pclass'].map(pclass_mapping)

# Display the resulting dataset with relevant encoded columns
print(encode_data[['Sex', 'Sex_Label', 'Embarked', 'Embarked_Q', 'Embarked_S', 'Embarked_C', 'Pclass', 'Pclass_Category']].head())
```
